```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scienceplots
plt.style.use(["science", "no-latex", "grid"])
```

```python
ys = pd.read_csv(r"..\..\bart_bo.csv", header=None)
ys_min = ys.cummin(axis=1)
```

```python
ys_min_avg = -ys_min.mean(axis=0)
ys_min_std = ys_min.std(axis=0)
x = np.arange(ys_min_avg.shape[0])
fig, ax = plt.subplots()
ax: plt.Axes
ax.plot(ys_min_avg, label="Alfalfa")
upper = ys_min_avg + ys_min_std
lower = ys_min_avg - ys_min_std
ax.fill_between(x, lower, upper, alpha=0.5)
ax.axhline(3.322, linestyle='--', color="C2", label="Maximum")

ax.set_ylabel("Black-box objective value")
ax.set_xlabel("Number of iterations")

ax.legend()
ax.set_xlim((0, 50))
fig.savefig("../../figs/bart_bo_runs.pdf")
```
