```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
x = 1.0
var = 0.1
step = stats.norm(0.0, var).rvs((10000,))
log_y = np.log(x) + step
y = np.exp(log_y)


plt.hist(y, bins=40, density=True)
lognormal = stats.lognorm(s=var, scale=x)
t = np.linspace(0, 10)
plt.plot(t, lognormal.pdf(t))
```

```python
# lognorm is not symmetric
X = 0.6931471824645996
Y = 0.2843815818924088
stats.lognorm.pdf(x=X, s=1.0, scale=Y), stats.lognorm.pdf(x=Y, s=1.0, scale=X)
```
